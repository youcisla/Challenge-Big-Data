```python
# 1. Convertir les 0 en pd.NA pour identifier correctement les valeurs manquantes
df['athlete_year_birth'] = df['athlete_year_birth'].replace(0, pd.NA)

# 2. Tenter d'imputer l'année de naissance si 'first_game_year' et 'age_at_first_game' sont disponibles
# Créer un masque pour les lignes où athlete_year_birth est NA et où les autres colonnes sont non-NA
mask_infer_birth_year = df['athlete_year_birth'].isna() & \
                        df['first_game_year'].notna() & \
                        df['age_at_first_game'].notna()

df.loc[mask_infer_birth_year, 'athlete_year_birth'] = \
    df.loc[mask_infer_birth_year, 'first_game_year'] - \
    df.loc[mask_infer_birth_year, 'age_at_first_game']

# 3. Calculer la médiane des années de naissance valides (non-NA)
median_birth_year = df['athlete_year_birth'].median()

# Remplir les NA restants avec la médiane
df['athlete_year_birth'] = df['athlete_year_birth'].fillna(median_birth_year).astype("Int64")

# 4. Recalculer 'age_at_first_game' pour qu'il soit cohérent avec la nouvelle 'athlete_year_birth'
df['age_at_first_game'] = df['first_game_year'] - df['athlete_year_birth']
df['age_at_first_game'] = df['age_at_first_game'].astype("Int64")

print("Vérification des valeurs manquantes après imputation pour athlete_year_birth et age_at_first_game :")
display(df[['athlete_year_birth', 'first_game_year', 'age_at_first_game']].isnull().sum())
```

```python
import pandas as pd
import ast

# 1️⃣ Lire le fichier HTML
file_path = "/content/drive/MyDrive/olympic_results.html"
tables = pd.read_html(file_path)
df_results = tables[0]

# 2️⃣ Nettoyage de base
df_results = df_results.drop_duplicates()

# 3️⃣ Gestion des valeurs manquantes
df_results['discipline_title'] = df_results['discipline_title'].fillna("Unknown")
df_results['event_title'] = df_results['event_title'].fillna("Unknown")
df_results['slug_game'] = df_results['slug_game'].fillna("Unknown")
df_results['participant_type'] = df_results['participant_type'].fillna("Unknown")
df_results['medal_type'] = df_results['medal_type'].fillna("None")
df_results['rank_equal'] = df_results['rank_equal'].fillna(False).astype(bool)
df_results['rank_position'] = pd.to_numeric(df_results['rank_position'], errors='coerce').fillna(0).astype(int)
df_results['country_name'] = df_results['country_name'].fillna("Unknown").str.strip()
df_results['country_code'] = df_results['country_code'].fillna("UNK")
df_results['country_3_letter_code'] = df_results['country_3_letter_code'].fillna("UNK")
df_results['athlete_url'] = df_results['athlete_url'].fillna("")
df_results['athlete_full_name'] = df_results['athlete_full_name'].fillna("")
df_results['value_unit'] = df_results['value_unit'].fillna("")
df_results['value_type'] = df_results['value_type'].fillna("")
df_results['athletes'] = df_results['athletes'].fillna("[]")  # pour pouvoir parser

# 4️⃣ Création de nouvelles colonnes pour ML

# Medal code
medal_map = {'GOLD':3, 'SILVER':2, 'BRONZE':1, 'None':0}
df_results['medal_code'] = df_results['medal_type'].map(medal_map)

# Num_athletes : compter le nombre d’athlètes
def count_athletes(x):
    try:
        if isinstance(x, str):
            x = ast.literal_eval(x)
        return len(x)
    except:
        return 0

df_results['num_athletes'] = df_results['athletes'].apply(count_athletes)

# Extraire l'année depuis slug_game
df_results['year'] = df_results['slug_game'].str.extract(r'(\d{4})').astype(int)

# Vérifications finales
print("Aperçu final :")
print(df_results.head())
print(df_results.info())
print("Valeurs manquantes par colonne :")
print(df_results.isnull().sum())

# Sauvegarde du fichier nettoyé
df_results.to_csv("/content/drive/MyDrive/olympic_results_clean.csv", index=False)
df_results.to_excel("/content/drive/MyDrive/olympic_results_clean.xlsx", index=False)
print("Fichier nettoyé des résultats sauvegardé avec succès !")
```

# jointure les fichiers

```python


#  Définir le chemin de chaque fichier
path_results = '/content/drive/MyDrive/olympic_results_clean .csv'
path_athletes = '/content/drive/MyDrive/olympic_athletes_clean.csv'
path_hosts = '/content/drive/MyDrive/olympic_hosts_clean .csv'
path_medals = '/content/drive/MyDrive/medals_clean.csv'

#  Lire les fichiers
df_results = pd.read_csv(path_results)
df_athletes = pd.read_csv(path_athletes)
df_hosts = pd.read_csv(path_hosts)
df_medals = pd.read_csv(path_medals)
```

```python
import pandas as pd

medals_type_count = df_medals.pivot_table(
    index=['year', 'slug_game', 'country_3_letter_code'],
    columns='medal_type',
    aggfunc='size',
    fill_value=0
).reset_index()

medals_type_count.columns.name = None
medals_type_count = medals_type_count.rename(columns={
    'GOLD': 'gold_medals',
    'SILVER': 'silver_medals',
    'BRONZE': 'bronze_medals'
})

medals_type_count['total_medals'] = (
    medals_type_count['gold_medals'] +
    medals_type_count['silver_medals'] +
    medals_type_count['bronze_medals']
)

athletes_per_country = df_results.groupby(
    ['year', 'slug_game', 'country_3_letter_code']
)['athlete_full_name'].nunique().reset_index()
athletes_per_country = athletes_per_country.rename(columns={'athlete_full_name': 'total_athletes'})

ages = df_results.merge(
    df_athletes[['athlete_full_name', 'athlete_year_birth']],
    on='athlete_full_name',
    how='left'
)

ages['age_at_games'] = ages['year'] - ages['athlete_year_birth']
ages = ages[ages['age_at_games'].between(10, 60)]

avg_age = ages.groupby(
    ['year', 'slug_game', 'country_3_letter_code']
)['age_at_games'].mean().reset_index()
avg_age = avg_age.rename(columns={'age_at_games': 'avg_age_athletes'})

history = medals_type_count[['country_3_letter_code', 'year', 'total_medals']].copy()
history = history.sort_values(['country_3_letter_code', 'year'], ascending=[True, True])
history = history.rename(columns={'total_medals': 'medals_in_current_year'})

df_hosts_simple = df_hosts[['game_slug', 'city', 'season', 'year', 'game_name']]

df_final = medals_type_count.merge(
    athletes_per_country,
    on=['year', 'slug_game', 'country_3_letter_code'],
    how='left'
)

df_final = df_final.merge(
    avg_age,
    on=['year', 'slug_game', 'country_3_letter_code'],
    how='left'
)

df_final = df_final.merge(
    history[['country_3_letter_code', 'year', 'medals_in_current_year']],
    on=['country_3_letter_code', 'year'],
    how='left'
)

df_hosts_simple = df_hosts[['game_slug', 'city', 'season', 'year', 'game_name']].copy()
country_mapping = df_medals[['country_name','country_3_letter_code']].drop_duplicates()
df_hosts_simple = df_hosts_simple.merge(country_mapping, left_on='city', right_on='country_name', how='left')


df_final = df_final.merge(
    df_hosts_simple,
    left_on=['year','slug_game'],
    right_on=['year','game_slug'],
    how='left'
)

df_final = df_final.sort_values(['country_3_letter_code_x', 'year'], ascending=[True,True])
df_final['cumulative_medals'] = df_final.groupby('country_3_letter_code_x')['medals_in_current_year'].cumsum().shift(1).fillna(0)
df_final['is_host'] = (df_final['country_3_letter_code_y'] == df_final['country_3_letter_code_x']).astype(int)

df_final = df_final.drop(columns=['country_name', 'country_3_letter_code_y', 'game_slug'])
df_final = df_final.reset_index(drop=True)

print(df_final.head())

print(df_final.columns)
```

### **Imputation des valeurs manquantes pour les athlètes**

Cette section gère les valeurs manquantes pour les années de naissance et l'âge au premier jeu. Elle essaie d'abord d'inférer les années de naissance à partir d'autres colonnes, puis utilise la médiane pour les valeurs restantes.

```python
# 1. Convertir les 0 en pd.NA pour identifier correctement les valeurs manquantes
df['athlete_year_birth'] = df['athlete_year_birth'].replace(0, pd.NA)

# 2. Tenter d'imputer l'année de naissance si 'first_game_year' et 'age_at_first_game' sont disponibles
# Créer un masque pour les lignes où athlete_year_birth est NA et où les autres colonnes sont non-NA
mask_infer_birth_year = df['athlete_year_birth'].isna() & \
                        df['first_game_year'].notna() & \
                        df['age_at_first_game'].notna()

df.loc[mask_infer_birth_year, 'athlete_year_birth'] = \
    df.loc[mask_infer_birth_year, 'first_game_year'] - \
    df.loc[mask_infer_birth_year, 'age_at_first_game']

# 3. Calculer la médiane des années de naissance valides (non-NA)
median_birth_year = df['athlete_year_birth'].median()

# Remplir les NA restants avec la médiane
df['athlete_year_birth'] = df['athlete_year_birth'].fillna(median_birth_year).astype("Int64")

# 4. Recalculer 'age_at_first_game' pour qu'il soit cohérent avec la nouvelle 'athlete_year_birth'
df['age_at_first_game'] = df['first_game_year'] - df['athlete_year_birth']
df['age_at_first_game'] = df['age_at_first_game'].astype("Int64")

print("Vérification des valeurs manquantes après imputation pour athlete_year_birth et age_at_first_game :")
display(df[['athlete_year_birth', 'first_game_year', 'age_at_first_game']].isnull().sum())
```

```python
import pandas as pd

file_path = "/content/drive/MyDrive/olympic_results.html"

# Lire toutes les tables du fichier HTML
tables = pd.read_html(file_path)

# Vérifier combien de tables ont été trouvées
print(f"Nombre de tables trouvées : {len(tables)}")

# Afficher un aperçu de la première table
df_results = tables[0]
print(df_results.head())
```

```python
import pandas as pd
import ast

# 1️⃣ Lire le fichier HTML
file_path = "/content/drive/MyDrive/olympic_results.html"
tables = pd.read_html(file_path)
df_results = tables[0]

# 2️⃣ Nettoyage de base
df_results = df_results.drop_duplicates()

# 3️⃣ Gestion des valeurs manquantes
df_results['discipline_title'] = df_results['discipline_title'].fillna("Unknown")
df_results['event_title'] = df_results['event_title'].fillna("Unknown")
df_results['slug_game'] = df_results['slug_game'].fillna("Unknown")
df_results['participant_type'] = df_results['participant_type'].fillna("Unknown")
df_results['medal_type'] = df_results['medal_type'].fillna("None")
df_results['rank_equal'] = df_results['rank_equal'].fillna(False).astype(bool)
df_results['rank_position'] = pd.to_numeric(df_results['rank_position'], errors='coerce').fillna(0).astype(int)
df_results['country_name'] = df_results['country_name'].fillna("Unknown").str.strip()
df_results['country_code'] = df_results['country_code'].fillna("UNK")
df_results['country_3_letter_code'] = df_results['country_3_letter_code'].fillna("UNK")
df_results['athlete_url'] = df_results['athlete_url'].fillna("")
df_results['athlete_full_name'] = df_results['athlete_full_name'].fillna("")
df_results['value_unit'] = df_results['value_unit'].fillna("")
df_results['value_type'] = df_results['value_type'].fillna("")
df_results['athletes'] = df_results['athletes'].fillna("[]")  # pour pouvoir parser

# 4️⃣ Création de nouvelles colonnes pour ML

# Medal code
medal_map = {'GOLD':3, 'SILVER':2, 'BRONZE':1, 'None':0}
df_results['medal_code'] = df_results['medal_type'].map(medal_map)

# Num_athletes : compter le nombre d’athlètes
def count_athletes(x):
    try:
        if isinstance(x, str):
            x = ast.literal_eval(x)
        return len(x)
    except:
        return 0

df_results['num_athletes'] = df_results['athletes'].apply(count_athletes)

# Extraire l'année depuis slug_game
df_results['year'] = df_results['slug_game'].str.extract(r'(\d{4})').astype(int)

# Vérifications finales
print("Aperçu final :")
print(df_results.head())
print(df_results.info())
print("Valeurs manquantes par colonne :")
print(df_results.isnull().sum())

# Sauvegarde du fichier nettoyé
df_results.to_csv("/content/drive/MyDrive/olympic_results_clean.csv", index=False)
df_results.to_excel("/content/drive/MyDrive/olympic_results_clean.xlsx", index=False)
print("Fichier nettoyé des résultats sauvegardé avec succès !")
```

### **Jointure et agrégation des différentes sources de données**

Cette section est cruciale. Elle charge les DataFrames nettoyés des médailles, athlètes, hôtes et résultats. Ensuite, elle agrège les données pour calculer le nombre de médailles par type, le nombre total d'athlètes, l'âge moyen des athlètes, et joint ces informations avec les données des hôtes pour créer un DataFrame final (`df_final`). Des caractéristiques supplémentaires comme les médailles cumulées et l'indicateur de pays hôte sont ajoutées.

```python
#  Définir le chemin de chaque fichier
path_results = '/content/drive/MyDrive/olympic_results_clean.csv'
path_athletes = '/content/drive/MyDrive/olympic_athletes_clean.csv'
path_hosts = '/content/drive/MyDrive/olympic_hosts_clean.csv'
path_medals = '/content/drive/MyDrive/medals_clean.csv'

#  Lire les fichiers
df_results = pd.read_csv(path_results)
df_athletes = pd.read_csv(path_athletes)
df_hosts = pd.read_csv(path_hosts)
df_medals = pd.read_csv(path_medals)
```

```python
import pandas as pd

medals_type_count = df_medals.pivot_table(
    index=['year', 'slug_game', 'country_3_letter_code'],
    columns='medal_type',
    aggfunc='size',
    fill_value=0
).reset_index()

medals_type_count.columns.name = None
medals_type_count = medals_type_count.rename(columns={
    'GOLD': 'gold_medals',
    'SILVER': 'silver_medals',
    'BRONZE': 'bronze_medals'
})

medals_type_count['total_medals'] = (
    medals_type_count['gold_medals'] +
    medals_type_count['silver_medals'] +
    medals_type_count['bronze_medals']
)

athletes_per_country = df_results.groupby(
    ['year', 'slug_game', 'country_3_letter_code']
)['athlete_full_name'].nunique().reset_index()
athletes_per_country = athletes_per_country.rename(columns={'athlete_full_name': 'total_athletes'})

ages = df_results.merge(
    df_athletes[['athlete_full_name', 'athlete_year_birth']],
    on='athlete_full_name',
    how='left'
)

ages['age_at_games'] = ages['year'] - ages['athlete_year_birth']
ages = ages[ages['age_at_games'].between(10, 60)]

avg_age = ages.groupby(
    ['year', 'slug_game', 'country_3_letter_code']
)['age_at_games'].mean().reset_index()
avg_age = avg_age.rename(columns={'age_at_games': 'avg_age_athletes'})

history = medals_type_count[['country_3_letter_code', 'year', 'total_medals']].copy()
history = history.sort_values(['country_3_letter_code', 'year'], ascending=[True, True])
history = history.rename(columns={'total_medals': 'medals_in_current_year'})

df_hosts_simple = df_hosts[['game_slug', 'city', 'season', 'year', 'game_name']]

df_final = medals_type_count.merge(
    athletes_per_country,
    on=['year', 'slug_game', 'country_3_letter_code'],
    how='left'
)

df_final = df_final.merge(
    avg_age,
    on=['year', 'slug_game', 'country_3_letter_code'],
    how='left'
)

df_final = df_final.merge(
    history[['country_3_letter_code', 'year', 'medals_in_current_year']],
    on=['country_3_letter_code', 'year'],
    how='left'
)

df_hosts_simple = df_hosts[['game_slug', 'city', 'season', 'year', 'game_name']].copy()
country_mapping = df_medals[['country_name','country_3_letter_code']].drop_duplicates()
df_hosts_simple = df_hosts_simple.merge(country_mapping, left_on='city', right_on='country_name', how='left')


df_final = df_final.merge(
    df_hosts_simple,
    left_on=['year','slug_game'],
    right_on=['year','game_slug'],
    how='left'
)

df_final = df_final.sort_values(['country_3_letter_code_x', 'year'], ascending=[True,True])
df_final['cumulative_medals'] = df_final.groupby('country_3_letter_code_x')['medals_in_current_year'].cumsum().shift(1).fillna(0)
df_final['is_host'] = (df_final['country_3_letter_code_y'] == df_final['country_3_letter_code_x']).astype(int)

df_final = df_final.drop(columns=['country_name', 'country_3_letter_code_y', 'game_slug'])
df_final = df_final.reset_index(drop=True)

print(df_final.head())

print(df_final.columns)
```
